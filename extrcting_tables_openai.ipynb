Extract PDFs from the ZIP File


KCS110000.zip

```python
from google.colab import files

uploaded = files.upload()
zip_file_name = next(iter(uploaded))
```

```python
import zipfile
import os

extraction_directory = '/content/pdf_files'
os.makedirs(extraction_directory, exist_ok=True)

# Extract the ZIP file
with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extractall(extraction_directory)

print("Extraction complete. PDFs are available in:", extraction_directory)
```

 Extract Tables from PDFs

```python
import fitz  # PyMuPDF

def extract_tables_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    tables = []

    for page in doc:
        # Assuming simple text extraction for now
        text = page.get_text("text")
        tables.append(text)  # Placeholder for your table extraction logic

    return tables

# Process each PDF
pdf_files = [f for f in os.listdir(extraction_directory) if f.endswith('.pdf')]
all_tables = []

for pdf_file in pdf_files:
    pdf_path = os.path.join(extraction_directory, pdf_file)
    tables = extract_tables_from_pdf(pdf_path)
    all_tables.extend(tables)
```

 Prepare for Querying and Query OpenAI

```python

import openai

# Ensure you have your OpenAI API key set securely
openai.api_key = 'enter you specific key'

# Adjusted to use the new ChatCompletion API for querying
for table_text in all_tables:
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",  # Use an appropriate model for your task
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": f"Analyze this table data: {table_text}"}
        ]
    )

    print(response['choices'][0]['message']['content'])
```
